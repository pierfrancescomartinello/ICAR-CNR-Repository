```python
filename = "data/scopus(1).csv"
scopus_data = pd.read_csv(filename)

scopus_data.columns
```

```python
# Delete unnecessary columns
scopus_data.drop(['Author full names', 'Author(s) ID', 'Source title','Link', 'Source', 'Open Access'], axis=1, inplace=True)
```

```python
scopus_data
```

Delete entries with:
- Document Type different than article, conference
- Publication Stage different than Final

```python
scopus_data = scopus_data.drop(scopus_data[
    #( scopus_data['DOI'] == "NaN") |
    ((scopus_data['Document Type'] != "Article") & (scopus_data['Document Type'] != "Conference paper") )|
    (scopus_data['Publication Stage'] != "Final")
    ].index)
```
